````python
import numpy as np
import matplotlib.pyplot as plt

# Задаем параметры
fs = 30000  # Частота дискретизации
f = 400    # Частота сигнала
duration = 1  # Длительность сигнала в секундах
t = np.arange(0, duration, 1/fs)  # Вектор времени

# Генерируем синусоидальный сигнал
A = 1  # Амплитуда сигнала
x = A * np.sin(2*np.pi*f*t)

# Добавляем Гауссов шум
sigma = 0.1  # Стандартное отклонение шума
noise = sigma * np.random.randn(len(x))
y = x + noise

# Визуализируем сигнал
plt.figure(figsize=(10, 6))
plt.plot(t, y)
plt.title("Синусоидальный сигнал с Гауссовым шумом")
plt.xlabel("Время, с")
plt.ylabel("Амплитуда")
plt.grid()
plt.show()
````

Объяснения:

1. ```import numpy as np```: Это импорт библиотеки NumPy, которая предоставляет мощные инструменты для работы с числовыми данными и математическими операциями.

2. ```import matplotlib.pyplot as plt```: Это импорт библиотеки Matplotlib, которая используется для визуализации данных и построения графиков.

3. ```# Задаем параметры```: В этом разделе мы определяем параметры для нашего сигнала и шума:
   - ```fs = 30000```: Частота дискретизации, то есть количество отсчетов сигнала в секунду.
   - ```f = 400```: Частота сигнала в Герцах.
   - ```duration = 1```: Длительность сигнала в секундах.
   - ```t = np.arange(0, duration, 1/fs)```: Создание вектора времени от 0 до длительности сигнала с шагом 1/fs (период дискретизации).

4. ```# Генерируем синусоидальный сигнал```: В этом блоке мы создаем синусоидальный сигнал с амплитудой 1 и частотой f:
   - ```A = 1```: Амплитуда сигнала.
   - ```x = A * np.sin(2*np.pi*f*t)```: Вычисление значений синусоидального сигнала.

5. ```# Добавляем Гауссов шум```: Здесь мы добавляем Гауссов шум к сигналу:
   - ```sigma = 0.1```: Стандартное отклонение шума.
   - ```noise = sigma * np.random.randn(len(x))```: Генерация Гауссова шума с заданным стандартным отклонением.
   - ```y = x + noise```: Сложение сигнала и шума, чтобы получить итоговый сигнал с шумом.

6. ```# Визуализируем сигнал```: В этом блоке мы используем Matplotlib для визуализации полученного сигнала:
   - ```plt.figure(figsize=(10, 6))```: Создание фигуры с заданным размером.
   - ```plt.plot(t, y)```: Построение графика сигнала `y` на временной оси `t`.
   - ```plt.title("Синусоидальный сигнал с Гауссовым шумом")```: Установка заголовка графика.
   - ```plt.xlabel("Время, с")```: Установка подписи оси X.
   - ```plt.ylabel("Амплитуда")```: Установка подписи оси Y.
   - ```plt.grid()```: Добавление сетки на график.
   - ```plt.show()```: Отображение построенного графика.

В целом, этот код моделирует синусоидальный сигнал с частотой 400 Гц, добавляет к нему Гауссов шум и визуализирует полученный сигнал с шумом.
